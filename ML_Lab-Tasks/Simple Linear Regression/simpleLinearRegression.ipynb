```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salary_dataset.csv")
df = df.reset_index(drop=True)
df
```

```python
Input = df.drop(["Salary","Unnamed: 0"], axis=1)
Input
```

```python
Output = df[["Salary"]]
Output
```

```python
np.shape(Input)
```

```python
np.shape(Output)
```

```python
plt.scatter(Input,Output)
plt.title("Salary predication")
plt.xlabel("year of the expreinces")
plt.ylabel("salary")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(Input,Output,test_size=0.4)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg
```

```python
reg.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
predict = reg.predict(X_test)
predict
```

```python
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='green', label='Testing Data')
plt.plot(X_test, predict, color='red', linewidth=3, label='Linear Regression')
plt.xlabel('exprience year')
plt.ylabel('salary')
plt.legend()
plt.show()
```

```python
reg.score(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```
