```python
# Import the library with an alias 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the Titanic dataset from a CSV file and store it in a DataFrame named 
titanic_data = pd.read_csv("titanic.csv")
titanic_data
```

```python
# Drop columns 'PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked' from the DataFrame 'df'
# The 'inplace=True' parameter modifies the original DataFrame instead of creating a new one
titanic_data.drop(["PassengerId","Name","Ticket","Cabin","Parch"],axis="columns",inplace=True)
```

```python
# converting the categorical to number for the 'Sex' column using one-hot encoding
titanic_data.replace({'Sex':{'male':0,'female':1},'Embarked':{'S':1,'Q':2,'C':3}},inplace =True)
titanic_data
```

```python
titanic_data
```

```python
# Check for columns with missing values (NaN)
titanic_data.isna().sum()
```

```python
# fill the nan values in the data set
titanic_data['Age'].fillna(titanic_data['Age'].mean(),inplace=True)
```

```python
titanic_data['Embarked'].fillna(titanic_data["Embarked"].mean(),inplace=True)
```

```python
# Check for columns with missing values (NaN)
titanic_data.isna().sum()
```

```python
# Create a new DataFrame 'inputs' excluding the 'Survived' column
inputs = titanic_data.drop(["Survived"],axis=1)
inputs
```

```python
# Create a Series 'target' containing the 'Survived' column
outputs = titanic_data['Survived']
outputs
```

```python
# Split the data into training and testing sets
# 70% of the data will be used for training, and 30% for testing
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(inputs,outputs,test_size=0.8,random_state=2)
```

```python
# Import the Gaussian Naive Bayes classifier from scikit-learn
from sklearn.naive_bayes import GaussianNB
```

```python
# Create a Gaussian Naive Bayes model
```

```python
model = GaussianNB()
model
```

```python
# Train the model on the training data
model.fit(X_train,y_train)
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
# test the data 
model.predict(X_test[:5])
```

```python
# Evaluate the model on the training data and print the accuracy score
model.score(X_train, y_train)
```

```python
# Evaluate the model on the testing data and print the accuracy score
model.score(X_test, y_test)
```

```python
# Import cross_val_score from scikit-learn for cross-validation
from sklearn.model_selection import cross_val_score

# Perform cross-validation and print the cross-validated scores
cross_val_score(GaussianNB(), X_train, y_train, cv=5)
```
