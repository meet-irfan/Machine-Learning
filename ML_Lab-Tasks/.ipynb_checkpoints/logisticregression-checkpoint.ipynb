```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
titanic_data = pd.read_csv('train.csv')
titanic_data
```

```python
# check the dataset rows and colomns
titanic_data.shape
```

```python
# to check the missing values of the dataset
titanic_data.isnull().sum()
```

```python
# now filling the missing values but cabin coloumn is missing 80% of the data and it is not in number form so we can not find the mean value of that coloumn thats why we should remove it
titanic_data = titanic_data.drop("Cabin", axis=1)
titanic_data
```

```python
#  the second coloumn is the age coloumn we fill it as the mean value of the age fill values
titanic_data['Age'].fillna(titanic_data['Age'].mean(),inplace =True)
titanic_data.isnull().sum()
```

```python
# the missing value of the embarked coloumn is and its output is s,c and q so than we find its mode means most repeated value
titanic_data['Embarked'].mode()[0]
```

```python
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0],inplace=True)
```

```python
# now check the missing values
titanic_data.isnull().sum()
```

```python
# how many people survived 
survived_counts = titanic_data['Survived'].value_counts()
survived_counts
```

```python
# Plot the bar chart
plt.bar(survived_counts.index, survived_counts.values, color=['red', 'green'])
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Distribution')
plt.xticks([0, 1], ['No', 'Yes'])  # Optional: to set custom labels for x-axis
plt.show()
```

```python
#  converting the catogracial columns 
titanic_data.replace({'Sex':{'male':0,'female':1},'Embarked':{'S':0,'C':1,'Q':2}},inplace=True)
titanic_data
```

```python
X = titanic_data.drop(["PassengerId","Survived","Name","Ticket"],axis=1)
X
```

```python
Y = titanic_data['Survived']
Y
```

```python
#  training and testing data
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.4,random_state=2)
```

```python
# model training 
model = LogisticRegression()
model
```

```python
model.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
y_predict = model.predict(X_test)
y_predict
```

```python
# checking the accuracy of the model
accuracy = model.score(X_train,y_train)
accuracy
```

```python
accuracy = model.score(X_test,y_test)
accuracy
```
